```python
data_new2 = data_new[['Hotel_Name','Hotel_Address','Nationality','Negative','Positive','Tags']]
data_new2
```

```python
data_new2['Tags'] = data_new2['Tags'].str.replace(" ", '')
data_new2['Tags'] = data_new2['Tags'].str.replace("'", '')
data_new2['Tags'] = data_new2['Tags'].str.replace("[", '')
data_new2['Tags'] = data_new2['Tags'].str.replace("]", '')
data_new2['Tags'] = data_new2['Tags'].str.replace(",", ' ')
```

```python
data_new2
```

```python
data_new2['tag']= data_new2['Hotel_Address']+data_new2['Nationality']+data_new2['Negative']+data_new2['Positive']+data_new2['Tags']
```

```python
data_new2.iloc[0].tag
```

```python
# data_new3 = pd.DataFrdata_new2[['Hotel_Name','Nationality','tag']]
data_new3 = pd.DataFrame(data_new2)
# data_new3 = data_new3[['Hotel_Name','Hotel_Address','Nationality','tag']]
# data_new3
```

```python
data_new3['tag'] = data_new3['tag'].apply(lambda x:x.lower())
```

```python
data_new3 = data_new3.reset_index(drop=True)
```

```python
data_new3 = data_new3[['Hotel_Name','Hotel_Address','Nationality','tag']]
# data_new3.iloc[0].tag
data_new3
```

```python
import nltk
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()

def stem(text):
  y = []

  for i in text.split():
    y.append(ps.stem(i))
  
  return " ".join(y)
# Xử lý những chữ cái có nghĩa giống nhau. VD: watched, watching -> watch
data_new3['tag'] = data_new3['tag'].apply(stem)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
cv_hotel = TfidfVectorizer(max_features=1124,stop_words='english')
```

```python
vectors_hotel = cv_hotel.fit_transform(data_new3['tag']).toarray()
vectors_hotel
```

```python
cv_hotel.fit_transform(data_new3['tag']).toarray().shape
```

```python
cv_hotel.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_hotel = cosine_similarity(vectors_hotel)
similarity_hotel
```

```python
sorted(list(enumerate(similarity_hotel[0])),reverse=True, key= lambda x:x[1])[1:10]
```

```python
def recommend(hotel):
  hotel_index = data_new3[data_new3['Hotel_Name'] == hotel].index[0]
  distances = similarity_hotel[hotel_index]
  hotel_list = sorted(list(enumerate(distances)),reverse=True, key= lambda x:x[1])[1:10]

  for i in hotel_list:
    print(data_new3.iloc[i[0]].Hotel_Name)
```

```python
recommend('Hilton Barcelona')
```
